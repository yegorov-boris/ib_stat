#### смещение среднего и медианы относительно a

150 наблюдений из N(a, sigma^2):

```python
def model_b(a, sigma, distr):
    mean = []
    med = []
    for _ in range(10):
        if distr == 0:
            n = 150
            s = np.random.normal(a, sigma, n)
        if distr == 1:
            n = 5
            s = np.random.normal(a + 5*sigma, n)
        if distr == 2:
            n = 155
            s0 = np.random.normal(a, sigma, 150)
            s1 = np.random.normal(a + 5*sigma, sigma, 5)
            s = np.concatenate((s0, s1))
        mean.append(np.mean(s))
        med.append(np.median(s))
    print('a = %f, sigma = %f' % (a, sigma))
    print('смещение выборочного среднего %f' % (np.mean(mean) - a))
    print('смещение выборочной медианы %f' % (np.mean(med) - a))
    s_mean = np.std(mean)
    s_med = np.std(med)
    print('дисперсия выборочного среднего %f' % ((n-1)*s_mean*s_mean/n))
    print('дисперсия выборочной медианы %f' % ((n-1)*s_med*s_med/n))
    print()
    
for a in [-10, -1, 0, 1, 10]:
    model_b(a, 1, 0)
    
for sigma in [0.01, 0.1, 1, 10, 100]:
    model_b(0, sigma, 0)
```

```python
def conf(theta, n, gamma):
    xn = np.max(np.random.uniform(low=0, high=theta, size=n))
    return (xn, xn/pow(1-gamma, 1/n))

def check_conf(theta, n, gamma):
    total = 1000
    m = 0
    for _ in range(total):
        (a, b) = conf(theta, n, gamma)
        if a < theta and theta < b:
            m += 1
    return m/total
```

```python
for theta in [1, 10, 100]:
    for gamma in [0.9, 0.95, 0.99]:
        for n in [100, 1000, 10000]:
            print('параметр theta = %f, уровень gamma = %f, объем выборки n = %d:' % (theta, gamma, n))
            print(conf(theta, n, gamma))
            print()
```

ширина интервала:

- пропорциональна параметру theta

- растет с увеличением уровня доверия gamma

- сходится к 0 с ростом объема выборки n

проверим, что интервал на самом деле доверительный:

возьмем 10000 выборок из U\[0, 1\], уровень доверия 0.9, объем выборки 10

```python
print(check_conf(1, 10, 0.9))
```

получили вероятность > 0.9

пример - стандартное распределение Коши C(0, 1)

функция распределения

$$F(x)=\frac{1}{\pi}arctg(x)+\frac{1}{2}$$

выборочное среднее из стандартного распределения Коши также является стандартным распределением Коши (и не является нормальным распределением):

характеристическая функция стандартного распределения Коши

$$\phi(t)=e^{-|t|}$$

выборочное среднее

$$\overset{-}{X}=\frac{1}{n}(X_1+...+X_n)$$

так как $$X_1,...,X_n$$ независимы, то характеристическая функция выборочного среднего

$$\phi_{\overset{-}{X}}(t)=(\phi(\frac{t}{n}))^n=(e^{-|\frac{t}{n}|})^n=e^{-|t|}=\phi(t)$$

так совпадают характеристические функции, то совпадают и распределения
